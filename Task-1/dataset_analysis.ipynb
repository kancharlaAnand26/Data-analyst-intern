```python
df.drop_duplicates(inplace=True)
```

```python
df.isnull().sum()
```

```python
df = df.dropna(subset=['Income'])
```

```python
df.isnull().sum()
```

```python
df.loc[:,'Education'] = df['Education'].str.strip().str.title()
df.loc[:,'Marital_Status'] = df['Marital_Status'].str.strip().str.title()
```

```python
df.head()
```

```python
df.loc[:,'Dt_Customer'] = pd.to_datetime(df['Dt_Customer'], errors='coerce')
```

```python
df.head()
```

```python
df.columns = df.columns.str.lower().str.replace(' ', '_')
```
